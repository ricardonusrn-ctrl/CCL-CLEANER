```python
from google.colab import files
import pandas as pd
import numpy as np
import os

# Install lasio library if not already installed
!pip install lasio
import lasio

print("Selecciona uno o varios archivos .las desde tu computador...")
uploaded = files.upload()

# Cut-off superior e inferior
cut_min = float(input("Ingresa el cut-off inferior (mínimo permitido): "))
cut_max = float(input("Ingresa el cut-off superior (máximo permitido): "))

for filename in uploaded.keys():
    print(f"\nProcesando archivo: {filename}")

    # Leer el archivo LAS
    las = lasio.read(filename)
    df = las.df()

    # Buscar curva CCL
    ccl_column = None
    for col in df.columns:
        if "CCL" in col.upper():
            ccl_column = col
            break

    if ccl_column is None:
        print(f"⚠ No se encontró curva CCL en {filename}. Se omite.")
        continue

    # -----------------------------------------------
    # ✅ CORRECCIÓN: mantener tendencia natural
    # -----------------------------------------------

    # Copia de la curva original
    ccl = df[ccl_column].copy()

    # Identificar valores fuera de rango
    mask_out = (ccl < cut_min) | (ccl > cut_max)

    # Sustituir valores fuera de rango por NaN (para permitir interpolación)
    ccl[mask_out] = np.nan

    # Interpolación lineal
    ccl_interp = ccl.interpolate(method="linear")

    # Rellenar NaN restantes (caso extremos: inicio/fin)
    ccl_interp = ccl_interp.fillna(method="bfill").fillna(method="ffill")

    # Reemplazar en DataFrame
    df[ccl_column] = ccl_interp

    # -----------------------------------------------
    # Actualizar curvas dentro del LAS
    # -----------------------------------------------
    for curve in las.curves:
        nombre = curve.mnemonic
        if nombre in df.columns:
            curve.data = df[nombre].values

    # -----------------------------------------------
    # Guardar archivo de salida
    # -----------------------------------------------
    base, ext = os.path.splitext(filename)
    output_name = f"{base}{ext}"

    las.write(output_name, version=2.0)
    print(f"✔ Archivo corregido generado: {output_name}")

    # Descargar archivo corregido
    files.download(output_name)

print("\n✅ Proceso completado con corrección suave (interpolación).")
```
