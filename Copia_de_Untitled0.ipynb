```python
from google.colab import files
import pandas as pd
import numpy as np
import os

# Install lasio library if not already installed
!pip install lasio
import lasio

print("Selecciona uno o varios archivos .las desde tu computador...")
uploaded = files.upload()

# Cut-off superior e inferior
cut_min = float(input("Ingresa el cut-off inferior (m√≠nimo permitido): "))
cut_max = float(input("Ingresa el cut-off superior (m√°ximo permitido): "))

# ---------------------------------------------------------
# ‚úÖ NUEVO: Elegir carpeta de destino
# ---------------------------------------------------------
print("\nSelecciona una carpeta de destino donde guardar los archivos corregidos...")
from google.colab import files
dest_folder = input("Escribe la ruta de la carpeta de destino (o deja vac√≠o para usar la carpeta actual): ")

# Si el usuario no escribe nada, se usa el directorio actual
if dest_folder.strip() == "":
    dest_folder = "."
else:
    # Crear carpeta si no existe
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)
        print(f"Carpeta creada: {dest_folder}")

print(f"üìÅ Los archivos corregidos se guardar√°n en: {os.path.abspath(dest_folder)}")
# ---------------------------------------------------------

for filename in uploaded.keys():
    print(f"\nProcesando archivo: {filename}")

    # Leer el archivo LAS
    las = lasio.read(filename)
    df = las.df()

    # Buscar curva CCL
    ccl_column = None
    for col in df.columns:
        if "CCL" in col.upper():
            ccl_column = col
            break

    if ccl_column is None:
        print(f"‚ö† No se encontr√≥ curva CCL en {filename}. Se omite.")
        continue

    # -----------------------------------------------
    # ‚úÖ CORRECCI√ìN DE PUNTOS FUERA DE RANGO
    # -----------------------------------------------
    ccl = df[ccl_column].copy()

    # Identificar valores fuera de rango
    mask_out = (ccl < cut_min) | (ccl > cut_max)

    # Sustituir valores fuera de rango por NaN para interpolaci√≥n
    ccl[mask_out] = np.nan

    # Interpolaci√≥n lineal
    ccl_interp = ccl.interpolate(method="linear")

    # Rellenar NaN restantes
    ccl_interp = ccl_interp.fillna(method="bfill").fillna(method="ffill")

    # Reemplazar curva en DataFrame
    df[ccl_column] = ccl_interp

    # -----------------------------------------------
    # Actualizar curvas dentro del LAS
    # -----------------------------------------------
    for curve in las.curves:
        nombre = curve.mnemonic
        if nombre in df.columns:
            curve.data = df[nombre].values

    # -----------------------------------------------
    # Guardar archivo en la carpeta seleccionada
    # -----------------------------------------------
    base, ext = os.path.splitext(filename)
    output_name = f"{base}{ext}"
    output_path = os.path.join(dest_folder, output_name)

    las.write(output_path, version=2.0)
    print(f"‚úî Archivo corregido generado: {output_path}")

    # Descargar archivo corregido
    files.download(output_path)

print("\n‚úÖ Proceso completado con correcci√≥n suave (interpolaci√≥n).")
```
